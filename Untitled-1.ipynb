```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
M = 1.0  # Mass parameter in units where G = c = 1
r_horizon = 2 * M  # Event horizon radius

# Range of r values: slightly outside and inside the horizon
r_outside = np.linspace(2.01 * M, 5 * M, 500)
r_inside = np.linspace(0.01, 1.99 * M, 500)

# Generate t(r) for radial light rays with positive and negative slopes outside and inside the horizon
def t_outside(r, sign):
    return sign * r + sign * 2 * M * np.log(np.abs(r - 2 * M))

def t_inside(r, sign):
    return sign * r + sign * 2 * M * np.log(np.abs(r - 2 * M))

# Initialize the plot
plt.figure(figsize=(8, 6))
plt.axvline(r_horizon, color="black", linestyle="--", linewidth=2)
plt.text(r_horizon + 0.05, 0, "$r=2M$", ha="left", va="center", fontsize=14,
         bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.3'))



# Plot light rays outside the horizon
for offset in np.linspace(-10, 10, 20):
    plt.plot(r_outside, t_outside(r_outside, +1) + offset, color="blue", alpha=0.4)
    plt.plot(r_outside, t_outside(r_outside, -1) + offset, color="blue", alpha=0.4)
# Plot light rays inside the horizon
for offset in np.linspace(-10, 10, 20):
    plt.plot(r_inside, t_inside(r_inside, +1) + offset, color="blue", alpha=0.4)
    plt.plot(r_inside, t_inside(r_inside, -1) + offset, color="blue", alpha=0.4)

# Labels and limits
plt.xlabel("$r$")
plt.ylabel("$t$")
plt.xlim(0, 5 * M)
plt.ylim(-10, 10)
plt.title("Radial light rays for the Schwarzschild geometry")
plt.grid(False)
# Save as PDF
plt.savefig("radial_light_rays.pdf", format="pdf")

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define psi and xi range
psi = np.linspace(0, np.pi, 500)
xi = np.linspace(-np.pi, np.pi, 500)
Psi, Xi = np.meshgrid(psi, xi)

# Define expressions for r and t
r = 0.5 * (np.tan(0.5 * (Psi + Xi)) - np.tan(0.5 * (Psi - Xi)))
t = 0.5 * (np.tan(0.5 * (Psi + Xi)) + np.tan(0.5 * (Psi - Xi)))

# Apply a mask to only include points within the diamond region |psi| + |xi| <= pi
mask = np.abs(Psi) + np.abs(Xi) <= np.pi
r = np.where(mask, r, np.nan)  # Mask r outside the diamond region
t = np.where(mask, t, np.nan)  # Mask t outside the diamond region

# Set up the plot
plt.figure(figsize=(6, 8))

# Define a range of constant r and t values to plot
r_levels = np.linspace(-3.14, 3.14, 9)   # Adjust these as needed for different contour levels
t_levels = np.linspace(0, 3.14, 9)       # Adjust for the range of t you want

# Plot the contours for constant r
contours_r = plt.contour(Psi, Xi, r, levels=r_levels, colors='royalblue', alpha=0.6, linestyles='solid')

# Plot the contours for constant t
contours_t = plt.contour(Psi, Xi, t, levels=t_levels, colors='darkred', alpha=0.6, linestyles='solid')

# Add straight lines that form the boundaries of the diamond shape
plt.plot(psi, np.pi - psi, 'k--', linewidth=1, alpha=0.7)
plt.plot(psi, psi - np.pi, 'k--', linewidth=1, alpha=0.7)

# Add labels at the corners
plt.text(np.pi, 0, "$i^0$", ha="right", va="center", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.text(0, -np.pi, "$i^-$", ha="center", va="center", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.text(0, np.pi, "$i^+$", ha="center", va="center", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Add labels to each surface
plt.text(0.2, 0.2, "$r=0$", ha="center", va="bottom", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.text(np.pi/2+0.2, np.pi / 2, "$\mathcal{I}^+$", ha="left", va="center", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.text(np.pi / 2+0.2, -np.pi/2, "$\mathcal{I}^-$", ha="center", va="top", fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Set up plot labels and grid
plt.xlabel(r"$\psi$")
plt.ylabel(r"$\xi$")
plt.title("Penrose Diagram of Minkowski Spacetime")
plt.grid(True, alpha=0.3)
plt.savefig("Minko_penrose.pdf", format="pdf")
plt.show()
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# Define r value (constant value for which we want the contour)
r_value = 1.0

# Define the equation to solve for xi in terms of psi
def equation(xi, psi, r):
    return np.tan(0.5 * (psi + xi)) - np.tan(0.5 * (psi - xi)) - r

# Define psi range
psi_vals = np.linspace(-np.pi, np.pi, 500)
xi_vals = []

# Solve for xi for each psi
for psi in psi_vals:
    xi_solution = fsolve(equation, x0=0, args=(psi, r_value))
    xi_vals.append(xi_solution[0])

# Plot psi vs xi for constant r
plt.plot(psi_vals, xi_vals, label=f'r = {r_value}')
plt.xlabel("psi")
plt.ylabel("xi")
plt.title("Line of Constant r in psi-xi Coordinates")
plt.legend()
plt.grid(True)
plt.show()
```
